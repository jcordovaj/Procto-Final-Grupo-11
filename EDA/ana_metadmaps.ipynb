```python
category=google_metadata['category'].apply(pd.Series).reset_index() #desanidar cetegorias
```

```python
category.head(2)
```

```python
categorymelt = category.melt(id_vars="index",var_name="category",value_name="catego")
```

```python
categorymelt.head()
```

```python
categorymelt.shape
```

```python
categ_uni=categorymelt['catego'].drop_duplicates()
```

```python
categ_uni.head() #lista con categorias unicas para archivo 1 de metadata
```

```python
categ_uni.to_csv('categ.csv')
```
