```python
# azureml-core of version 1.0.72 or higher is required
# azureml-dataprep[pandas] of version 1.1.34 or higher is required
from azureml.core import Workspace, Dataset

subscription_id = '2dc22500-40aa-4abb-bb2b-6426e3234bad'
resource_group = 'henryyelp'
workspace_name = 'henry_maps_yelp'

workspace = Workspace(subscription_id, resource_group, workspace_name)

dataset = Dataset.get_by_name(workspace, name='review')
df =dataset.to_pandas_dataframe() #lectura de archivo review de Yelp, cargado en azure
```

```python
df.shape
```

```python
df.head()
```

```python
review_yelp_pruebav=df.loc[0:1000] # se crea una muestra con 1000 registros para review
```

```python
review_yelp_pruebav.to_csv('review_yelp_prueba.csv',index=False)
```

```python
review_yelp_pruebav.head()
```

```python
review_yelp_pruebav.to_json('review_yelp_pruebav2.json',index=False,lines=True,orient='table')
```
